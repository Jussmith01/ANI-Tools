```python
from rdkit import Chem
import pyaniasetools as pya

from ase_interface import ANIENS,ensemblemolecule
import hdnntools as hdt
import numpy as np

%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt

import time
```

```python
nets = dict()

ntdir = '/home/jsmith48/scratch/ANI-2x_retrain/train/ANI-2x_v62518/'
#ntdir = '/home/jsmith48/scratch/transfer_learning/train_ens_DFTTZ/'
nets['ANI-1x(DFTDZ)']= {'cns' : ntdir + 'train0/rHCNOSFCl-5.2R_16-3.8A_a4-8.params',
                                 'sae' : ntdir + 'train0/sae_linfit.dat',
                                 'nnf' : ntdir + 'train',
                                 'Nn'  : 8}
```

```python
ens = ensemblemolecule(nets['ANI-1x(DFTDZ)']['cns'], nets['ANI-1x(DFTDZ)']['sae'], nets['ANI-1x(DFTDZ)']['nnf'], nets['ANI-1x(DFTDZ)']['Nn'], 7)
```

```python
mol = Chem.MolFromMolFile('/data/cdever01/cys/dimer/md_dicysteine_dimer.mol', removeHs=False)
#mol = Chem.MolFromMolFile('/data/cdever01/cys/dimer/dicysteine_dimer.mol', removeHs=False)

ts = pya.ani_tortion_scanner(ens, fmax=1.0, printer=True)

torsions = {'Phi':[1,2,12,13],'Psi':[0,1,2,12]}

st = time.time()
p,e,s = ts.scan_tortion(mol, torsions, 5.0, 74)
print(time.time()-st)
```

```python
from scipy.ndimage import zoom 

n_zoom = 8
data_x = zoom(p[:,:,0],n_zoom,order=1)
data_y = zoom(p[:,:,1],n_zoom,order=1)
data_z = zoom(e-e.min(),n_zoom,order=1)

fig = plt.figure(figsize=(18,12))
plt.style.use('seaborn-white')
contours = plt.contour(data_x, data_y, data_z, 25, colors='black')
plt.clabel(contours, inline=True, fontsize=12)

im1 = plt.imshow(data_z.T, extent=[data_x.min(), data_x.max(), data_y.min(), data_y.max()], origin='lower',
           cmap='jet', alpha=1.0, interpolation='gaussian')

print(im1)
plt.xlabel('Phi',fontsize=22)
plt.ylabel('Psi',fontsize=22)

CB=plt.colorbar()
CB.set_label('Delta E (kcal/mol)', fontsize=16)

#plt.axis(aspect='image');
print('Done')
```

```python
angle=[60.0, -90.0]
def get_angle_pos(find_idx, rho):
    ids=[]
    for i,ps in enumerate(rho):
        for j,pe in enumerate(ps):
            if np.allclose(np.array(find_idx), pe):
                ids.append((i, j))
    return ids
```
