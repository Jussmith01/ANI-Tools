```python
# Set NC
aens = aat.anicrossvalidationconformer(cns, sae, nnf, Nn, [0])
```

```python
Eal=[]
dEal=[]
Emin=[]

Na = []
adl = pyt.anidataloader(ts)
for i,data in enumerate(adl):

    # Get reference data
    X = np.array(data['coordinates'],dtype=np.float32)
    S = data['species']
    E = hdt.hatokcal*data['energies']
    Na.append(len(S))

    # Define Eref
    Eref = E

    # Compute energies and forces
    Ea1_t,Sa1 = aens.compute_energy_conformations(X,S)
    
#     ae = np.empty((Nn,E.size,len(S)),dtype=np.float64)
#     for j,nc in enumerate(aens.ncl):
#         ae[j] = nc.aenergies().copy()
    
#     print(hdt.hatokcal*ae[:,0,8])
    
#     for s in set(S):
#         sidx = np.where(np.array(S) == s)
#         #sae = ae[:,:,sidx[0]].reshape((Nn,-1))
#         sigae = hdt.hatokcal*np.std(sae,axis=0)
#         menae = hdt.hatokcal*np.mean(np.abs(sae))
#         sumae = hdt.hatokcal*np.sum(np.mean(sae,axis=0))
#         print(s,sigae.max(),menae,sumae/sidx[0].size)
    
    print('Computing:',i,len(S),data['path'],'Nm:',E.size,'Na:',len(S),'Err:',hdt.calculatemeanabserror(Eref,Ea1_t))
    
    # Define NN shifted energies
    Eal1 = Ea1_t # TL

    # Define NN E distmats
    dEal1 = hdt.calculatedmat(Ea1_t)

    # Store E mins
    Emin.append(np.array([[Eref.min(),Eal1.min()]]))

    # Store E and dE
    Eal.append(np.vstack([Eref,Eal1,Sa1]))
    dEal.append(np.vstack([hdt.calculatedmat(Eref),dEal1]))
    #print(dEal[-1].shape)
        
Eal = np.hstack(Eal)
dEal = np.hstack(dEal)
#print(dEal.shape,dEal)
Na = np.array(Na)
Emin=np.vstack(Emin)
```
